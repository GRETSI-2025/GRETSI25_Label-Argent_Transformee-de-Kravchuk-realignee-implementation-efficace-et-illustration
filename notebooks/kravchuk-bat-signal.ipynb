#### Load a bat signal and display it

from https://tftb.nongnu.org/demos.html, section $\mathsf{Animals:\, the \, bat \, call}$

```python
samplerate, data = wavfile.read('../data/bat.wav')
```

#### Append one point to have a signal of N + 1 points with N even and normalize the signal

```python
signal           = np.array(data,dtype = 'float64')
signal           = np.append(signal, 0)
signal          /= np.linalg.norm(signal,ord=2)
```

#### Definition of the time samples from the sample rate

```python
time_t           = samplerate ** (-1) * np.arange(len(signal))
```

#### Display the bat signal

```python
_                = display_signal(signal,time_t,yticks=False)

#plt.savefig('../figures/bat.eps')
```

#### Compute the Kravchuk transform of the bat signal

```python
Fs, thetas, phis = the_new_transform(signal)
```

#### Display the Kravchuk transform 

```python
planar_display(np.exp(np.abs(Fs)),[],[],thetas,phis)
```

```python
rsignal = rotate_signal(signal, 4*np.pi/3, -2*np.pi/3)

Rs, thetas, phis = the_new_transform(rsignal)
```

```python
spherical_display(np.exp(np.abs(Rs)),[],[],thetas,phis)
```

#### Compute the Short-Time Fourier Transform

```python
Vs, time_t, fs  = the_stft_transform(signal)
```

#### Display the Gaussian spectrogram
